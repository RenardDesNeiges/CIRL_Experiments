```python
import numpy as np
import numpy.random as rd
import jax
from jax import numpy as jnp
from jax.numpy import linalg as jla
import matplotlib.pyplot as plt
from algs.projs import euclidean_l1ball
from typing import Tuple
from itertools import accumulate
from einops import rearrange
Point = Tuple[int,int]
Goal = Tuple[Point,float]

def flatten(v):
    return jnp.reshape(v,(list(accumulate(v.shape,lambda x,y:x*y))[-1],))
```

### Reniy Entropy regularizer

Defined as : 
$$
H_\alpha(\pi) = \frac{\alpha}{1-\alpha} \cdot \log \| \pi \|_\alpha
$$

```python
def renyi_h2(pi,alpha):
    return (alpha/(1-alpha)) * jnp.log(jla.norm(pi,alpha))
```

```python
n = 2
constraints = [
    [0,0.2,0.,0],
    [0,0.2,0.,0],]
def cost_matrix(a,c):
    return c[a]
b = np.ones((len(constraints)))*2.0
Psi = np.array(  
        [[[ cost_matrix(a,c) 
            for _  in range(1)             ] 
            for a  in range(n)             ] 
            for c  in constraints         ],dtype=np.float64) * 0.2
Bandit = Complete(1,n,1.,b=b,Psi=Psi,rmax=2.) # Bandits are just one state MDPs
Bandit.R = np.array([1.4, 6.2])
Bandit.R = euclidean_l1ball(Bandit.R) # make sur reward is in reward class
```

### Implementing a saddle-point forward Bandit solver

We want to solve the MDP by solving 
$$
\min_{\zeta \succcurlyeq 0} \max_{\pi \in  \Delta_a} r^T \pi + \beta H_2(\pi) - \zeta^T(b-\Psi \pi),
$$
where :
$$
\text{obj}(\pi,\zeta) = r^T \pi + \beta H_2(\pi) - \zeta^T(b-\Psi \pi).
$$

```python
def obj(pi, zeta, r, beta, b, Psi):
    Psi_t = rearrange(Psi,'c a s -> (a s) c').transpose()
    return jnp.dot(flatten(r),flatten(pi)) \
        + beta * renyi_h2(flatten(pi),2) \
            + jnp.dot(flatten(zeta),flatten(b-(Psi_t@pi.transpose()).transpose()))
```

### Projected Expected Gradient GDA implementation (for the forward bandit problem)

```python
def forward_bandit_EG( Bandit: MarkovDecisionProcess,
                        iterations:int = 60,
                        eta:float = 1e-1,  # set the learning rate
                        beta:float = 0.2,  # set the value of the regularizer constant
                        ):
    def eval_grads(pi,zeta,beta,Bandit):
        val = obj(pi,zeta,Bandit.R,beta,Bandit.b,Bandit.Psi)
        grad_zeta = jax.grad(obj,1)(pi,zeta,Bandit.R,beta,Bandit.b,Bandit.Psi)
        grad_pi = jax.grad(obj,0)(pi,zeta,Bandit.R,beta,Bandit.b,Bandit.Psi)
        return val, grad_pi, grad_zeta


    zeta_t = rd.random(Bandit.b.shape) # initialize dual variable (for the constraints)
    pi_t = euclidean_simplex(rd.random((1,n))) # initialize policy somewhere in the simplex
    vals = []
    pis = []
    zetas = []
    normfs = []
    for _ in range(iterations):
        _, grad_pi, grad_zeta = eval_grads(pi_t,zeta_t,beta,Bandit)
        zeta_t_half = euclidian_pos_orthant( zeta_t-eta*grad_pi )
        pi_t_half = euclidean_simplex( pi_t+eta*grad_pi )
        val, grad_pi, grad_zeta = eval_grads(pi_t_half,zeta_t_half,beta,Bandit)
        zeta_t = euclidian_pos_orthant( zeta_t-eta*grad_pi )
        pi_t = euclidean_simplex( pi_t+eta*grad_pi )
        vals += [val]; pis += [pi_t]; zetas += [zeta_t]; normfs += [jla.norm(jnp.concatenate([eta*grad_pi,eta*grad_zeta],1),2)]
    return pi_t, zeta_t, pis, vals, zetas, normfs

pi_forward, zeta_forward, pis, vals, zetas, normfs = forward_bandit_EG(Bandit)
```

### Inverse Bandits

We want to solve the IRL-bandit problem through:
$$
\min_{\zeta \succcurlyeq 0, r \in \mathcal{R}_{L_1}} \max_{\pi \in  \Delta_a} r^T (\pi-\pi^E) + \beta H_2(\pi) - \zeta^T(b-\Psi \pi),
$$
where :
$$
\text{obj}(\pi,\zeta, r) = r^T  (\pi-\pi^E)  + \beta H_2(\pi) - \zeta^T(b-\Psi \pi).
$$

```python
def irl_obj(pi, zeta, r, beta, b, Psi,pi_E):
    Psi_t = rearrange(Psi,'c a s -> (a s) c').transpose()
    return jnp.dot(flatten(r),flatten(pi)-flatten(pi_E)) \
        + beta * renyi_h2(flatten(pi),2) \
            + jnp.dot(flatten(zeta),flatten(b-(Psi_t@pi.transpose()).transpose()))
```

```python
def inverse_bandit_EG( Bandit: MarkovDecisionProcess,
                        pi_E:np.ndarray,
                        iterations:int = 60,
                        eta:float = 1e-1,  # set the learning rate
                        beta:float = 0.2,  # set the value of the regularizer constant
                        ):
    def eval_grads(pi,zeta,r,beta,Bandit):
        val = obj(pi,zeta,r,beta,Bandit.b,Bandit.Psi)
        grad_r = jax.grad(irl_obj,2)(pi,zeta,r,beta,Bandit.b,Bandit.Psi,pi_E)
        grad_zeta = jax.grad(irl_obj,1)(pi,zeta,r,beta,Bandit.b,Bandit.Psi,pi_E)
        grad_pi = jax.grad(irl_obj,0)(pi,zeta,r,beta,Bandit.b,Bandit.Psi,pi_E)
        return val, grad_pi, grad_zeta, grad_r


    zeta_t = rd.random(Bandit.b.shape) # initialize dual variable (for the constraints)
    r_t = euclidean_l1ball(rd.random((1,n))) # initialize reward somewhere in the L1 ball (which is the reward class in this example)
    pi_t = euclidean_simplex(rd.random((1,n))) # initialize policy somewhere in the simplex
    vals = []
    pis = []
    pis_avg = []
    rs = []
    rs_avg = []
    zetas = []
    zetas_avg = []
    normfs = []
    for _ in range(iterations):
        _, grad_pi, grad_zeta, grad_r = eval_grads(pi_t,zeta_t,r_t,beta,Bandit)
        zeta_t_half = euclidian_pos_orthant( zeta_t-eta*grad_pi )
        r_t_half = euclidean_l1ball( r_t-eta*grad_r )
        pi_t_half = euclidean_simplex( pi_t+eta*grad_pi )
        val, grad_pi, grad_zeta, grad_r = eval_grads(pi_t_half,zeta_t_half,r_t_half,beta,Bandit)
        zeta_t = euclidian_pos_orthant( zeta_t-eta*grad_pi )
        r_t = euclidean_l1ball( r_t-eta*grad_r )
        pi_t = euclidean_simplex( pi_t+eta*grad_pi )
        vals += [val]; pis += [pi_t]; zetas += [zeta_t]; rs += [r_t]; normfs += [jla.norm(jnp.concatenate([eta*grad_pi,eta*grad_zeta],1),2)]
        pis_avg += [np.mean(np.array(pis),0)[0]]; zetas_avg += [np.mean(np.array(zetas),0)[0]]; rs_avg += [np.mean(np.array(rs),0)]
    return pi_t, zeta_t, pis, vals, zetas, rs, normfs, pis_avg, zetas_avg, rs_avg

_,_, pis, vals, zetas, rs, normfs, pis_avg, zetas_avg, rs_avg = inverse_bandit_EG(Bandit,pi_forward,200,eta=1e-1)
```
